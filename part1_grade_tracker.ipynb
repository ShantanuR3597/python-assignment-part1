## Task 1 — Data Parsing & Profile Cleaning

```python
# Raw student data (given in question)
raw_students = [
    {"name": " ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# Creating a new list to store cleaned data
cleaned_students = []

# Looping through each student to clean data
for student in raw_students:

    # Remove extra spaces and fix name casing
    name = student["name"].strip().title()

    # Convert roll number from string to integer
    roll = int(student["roll"])

    # Convert marks string into list of integers
    marks = []
    for m in student["marks_str"].split(","):
        marks.append(int(m.strip()))

    # Storing cleaned data in a new dictionary
    cleaned_students.append({
        "name": name,
        "roll": roll,
        "marks": marks
    })

# Display cleaned data
print("Cleaned Student Data:\n")
for student in cleaned_students:
    print(student)
```

```python
# Checking if names contain only alphabets

print("\nName Validation:\n")

for student in cleaned_students:

    name_parts = student["name"].split()

    # Check each part of name (first/last)
    if all(part.isalpha() for part in name_parts):
        print(f"{student['name']} → Valid")
    else:
        print(f"{student['name']} → Invalid")
```

```python
# Printing student details in a clean format

print("\nStudent Profiles:\n")

for student in cleaned_students:
    print("=" * 30)
    print(f"Student : {student['name']}")
    print(f"Roll No : {student['roll']}")
    print(f"Marks   : {student['marks']}")
    print("=" * 30)
```

```python
# Finding student with roll number 103

for student in cleaned_students:
    if student["roll"] == 103:

        print("\nSpecial Output:\n")

        # Print name in uppercase
        print(student["name"].upper())

        # Print name in lowercase
        print(student["name"].lower())
```

## Task 2 — Marks Analysis Using Loops & Conditionals

```python
# Given data
student_name = "Ayesha Sharma"

subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]
```

```python
print("\nSubject-wise Performance:\n")

# Loop through subjects and marks together
for i in range(len(subjects)):

    subject = subjects[i]
    score = marks[i]

    # Assign grade based on score
    if score >= 90:
        grade = "A+"
    elif score >= 80:
        grade = "A"
    elif score >= 70:
        grade = "B"
    elif score >= 60:
        grade = "C"
    else:
        grade = "F"

    print(f"{subject} : {score} → {grade}")
```

```python
# Total marks
total_marks = sum(marks)

# Average marks
average_marks = total_marks / len(marks)

# Highest scoring subject
max_marks = max(marks)
max_index = marks.index(max_marks)
highest_subject = subjects[max_index]

# Lowest scoring subject
min_marks = min(marks)
min_index = marks.index(min_marks)
lowest_subject = subjects[min_index]

print("\nSummary:\n")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {round(average_marks, 2)}")
print(f"Highest: {highest_subject} ({max_marks})")
print(f"Lowest: {lowest_subject} ({min_marks})")
```

```python
# Allow user to add more subjects

print("\nAdd New Subjects (type 'done' to stop):\n")

while True:

    subject = input("Enter subject name: ")

    # Stop condition
    if subject.lower() == "done":
        break

    marks_input = input("Enter marks (0-100): ")

    # Validate input
    if not marks_input.isdigit():
        print("Invalid input! Marks must be a number.")
        continue

    marks_value = int(marks_input)

    if marks_value < 0 or marks_value > 100:
        print("Marks should be between 0 and 100.")
        continue

    # Add valid data
    subjects.append(subject)
    marks.append(marks_value)

    print(f"{subject} added successfully!")
```

```python
# Updated average after adding subjects
new_total = sum(marks)
new_average = new_total / len(marks)

print("\nUpdated Summary:\n")
print(f"Subjects: {subjects}")
print(f"Marks: {marks}")
print(f"New Average: {round(new_average, 2)}")
```

## Task 3 — Class Performance Summary

```python
# Given class data
class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]
```

```python
print("\nClass Report:\n")

# Counters for pass/fail
pass_count = 0
fail_count = 0

# Store averages for later calculations
averages = []

# Store topper info
topper_name = ""
topper_avg = 0

# Header
print("Name              | Average | Status")
print("-" * 40)

# Loop through each student
for student in class_data:

    name = student[0]
    marks = student[1]

    # Calculate average
    avg = sum(marks) / len(marks)
    avg = round(avg, 2)

    averages.append(avg)

    # Assign pass/fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # Check for topper
    if avg > topper_avg:
        topper_avg = avg
        topper_name = name

    # Print formatted row
    print(f"{name:<18} | {avg:<7} | {status}")
```

```python
# Calculate class average
class_average = sum(averages) / len(averages)

print("\nSummary:\n")

print(f"Passed Students: {pass_count}")
print(f"Failed Students: {fail_count}")

print(f"Class Topper: {topper_name} ({topper_avg})")

print(f"Class Average: {round(class_average, 2)}")
```

## Task 4 — String Manipulation Utility

```python
# Given essay
essay = " python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science. "

# Remove extra spaces from start and end
clean_essay = essay.strip()

print("Clean Essay:\n")
print(clean_essay)
```

```python
# Convert to title case
title_case_essay = clean_essay.title()

print("\nTitle Case:\n")
print(title_case_essay)
```

```python
# Count how many times 'python' appears
count_python = clean_essay.count("python")

print("\nCount of 'python':", count_python)
```

```python
# Replace all occurrences
replaced_essay = clean_essay.replace("python", "Python 🐍")

print("\nReplaced Essay:\n")
print(replaced_essay)
```

```python
# Split essay into sentences
sentences = clean_essay.split(". ")

print("\nSentences List:\n")
print(sentences)
```

```python
print("\nNumbered Sentences:\n")

for i in range(len(sentences)):

    sentence = sentences[i]

    # Add full stop if missing
    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{i+1}. {sentence}")
```
